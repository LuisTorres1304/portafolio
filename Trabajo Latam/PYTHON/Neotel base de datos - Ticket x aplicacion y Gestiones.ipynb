#### PARA LEER VARIOS ARCHIVOS TXT

```python
filas = []

# Itera sobre cada archivo en la carpeta
for archivo_nombre in os.listdir(carpeta):
    # Verifica si el archivo es un archivo de texto
    if archivo_nombre.endswith('.txt'):
        # Construye la ruta completa del archivo
        archivo_ruta = os.path.join(carpeta, archivo_nombre)

        # Abre el archivo en modo lectura
        with open(archivo_ruta, 'r') as archivo:
            # Lee cada línea del archivo
            for linea in archivo:
                # Divide la línea en columnas utilizando el tabulador como separador
                columnas = linea.strip().split('\t')

                # Agrega las columnas a la lista de filas
                filas.append(columnas)

# Convierte la lista de filas en un DataFrame
df = pd.DataFrame(filas)

# Imprime el DataFrame
df
```

```python
df.columns = df.iloc[0]

# Elimina la primera fila, ya que ahora es la columna de encabezados
df = df[1:]

# Reinicia los índices del DataFrame
df.reset_index(drop=True, inplace=True)

# Imprime el DataFrame después de cambiar los encabezados
df
```

### cuando quieres concantenar varios csv 

```python
import os

# Ruta de la carpeta que contiene los archivos CSV
ruta_carpeta = 'C:/Users/Usuario/Proyectos_Anaconda/WORK LR/Neotel/'

# Lista para almacenar los DataFrames individuales
dataframes = []

# Itera sobre los archivos en la carpeta
for archivo in os.listdir(ruta_carpeta):
    if archivo.endswith('.csv'):
        ruta_archivo = os.path.join(ruta_carpeta, archivo)
        
        # Lee el archivo CSV y agrega el DataFrame a la lista
        try:
            df = pd.read_csv(ruta_archivo, encoding='utf-8')
        except UnicodeDecodeError:
            print(f"Error al leer el archivo: {ruta_archivo}. Intentando con encoding latin-1.")
            df = pd.read_csv(ruta_archivo, encoding='latin-1')
            
        dataframes.append(df)

# Combina todos los DataFrames en uno solo
df_final = pd.concat(dataframes, ignore_index=True)

# Imprime el DataFrame resultante
df_final
```

```python
df_final.to_csv(f"C:/Users/Usuario/Proyectos_Anaconda/WORK LR/Neotel/union_nuevo2.csv", index=False, encoding= 'latin-1') # 
```
