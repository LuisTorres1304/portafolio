## LIMPIA DE LA COLUMNA REGISTRO

```python
reporte['Producto']= reporte['Producto'].str.replace('    ', ' ')
reporte['Producto']= reporte['Producto'].str.replace('   ', ' ')
reporte['Producto']= reporte['Producto'].str.replace('  ', ' ')
reporte['Producto'] = reporte['Producto'].str.title()
```

```python
reporte_SinPresentacion = reporte[~(reporte['Producto'].str.contains('- Vaso|- Jarra|- Copa| - Entera|- Medio| - Entero|- Botella|- Media'))]
```

```python
reporte_presentacion = reporte[reporte['Producto'].str.contains('- Vaso|- Jarra|- Copa| - Entera|- Medio| - Entero|- Botella|- Media')]
```

```python
reporte_presentacion['Presentación'] = reporte_presentacion['Producto'].str.split(' - ').str[1]
```

```python
reporte_presentacion['Presentación']= reporte_presentacion['Presentación'].str.replace(': Acholado','')
reporte_presentacion['Presentación']= reporte_presentacion['Presentación'].str.replace(' Frozen','')
reporte_presentacion['Presentación']= reporte_presentacion['Presentación'].str.replace(' Frosen','')
reporte_presentacion['Presentación']= reporte_presentacion['Presentación'].str.replace(' Acholado','')
reporte_presentacion['Presentación']= reporte_presentacion['Presentación'].str.replace(': Italia','')
reporte_presentacion['Presentación']= reporte_presentacion['Presentación'].str.replace(' Especial','')
reporte_presentacion['Presentación']= reporte_presentacion['Presentación'].str.replace(': Quebranta','')
reporte_presentacion['Presentación']= reporte_presentacion['Presentación'].str.replace(' Quebranta','')
```

```python
reporte_presentacion['Presentación'].value_counts()
```

```python
reporte2 = pd.concat([reporte_presentacion,reporte_SinPresentacion])
reporte2 = reporte2[reporte2['Cantidad'] >=1]
reporte2 = reporte2.sort_values('Fecha')
```

```python
reporte2['Producto']= reporte2['Producto'].str.replace('.','')
reporte2['Producto']= reporte2['Producto'].str.replace('-',' ')
reporte2['Producto']= reporte2['Producto'].str.replace('    ', ' ')
reporte2['Producto']= reporte2['Producto'].str.replace('   ', ' ')
reporte2['Producto']= reporte2['Producto'].str.replace('  ', ' ')
reporte2['Producto']= reporte2['Producto'].str.strip()
```

```python
reporte2['Producto']= reporte2['Producto'].str.replace('2Piso', '')
reporte2['Producto']= reporte2['Producto'].str.replace('2piso', '')
reporte2['Producto']= reporte2['Producto'].str.replace('2 Piso', '')
reporte2['Producto']= reporte2['Producto'].str.replace('PISO 2', '')
reporte2['Producto']= reporte2['Producto'].str.replace('PISO 1', '')
reporte2['Producto']= reporte2['Producto'].str.replace('Piso 1', '')
reporte2['Producto']= reporte2['Producto'].str.replace('1Piso', '')
reporte2['Producto']= reporte2['Producto'].str.replace('2Piso','')
reporte2['Producto']= reporte2['Producto'].str.replace('No', ' ')
reporte2['Producto']= reporte2['Producto'].str.title()
reporte2['Producto']= reporte2['Producto'].str.strip()
```

```python
reporte2['Producto']= reporte2['Producto'].str.replace('   ', ' ')
reporte2['Producto']= reporte2['Producto'].str.replace('  ', ' ')
reporte2['Producto'] = reporte2['Producto'].str.replace('Zucardi Q Malbec ,','Zucardi Q Malbec')
#reporte2['Producto'] = reporte2['Producto'].str.replace('Adicionales ','')
#reporte2['Producto'] = reporte2['Producto'].str.replace('Adicional ','Ads ')
reporte2['Producto'] = reporte2['Producto'].str.replace(' Ads ',' ')
reporte2['Producto'] = reporte2['Producto'].str.replace(' Ads ',' ')
```

```python
ax = reporte2.loc[(reporte2['Producto']=='Pechuga De Pollo')&(reporte2['Categoria']=='Adicionales')]
ay = reporte2.loc[~((reporte2['Producto']=='Pechuga De Pollo')&(reporte2['Categoria']=='Adicionales'))]
ax['Producto'] = ax['Producto'].str.replace('Pechuga De Pollo','Adicionales Pechuga De Pollo')
```

```python
primer_pedido = reporte3.groupby(['Fecha','Hora','Mesa','Cliente','Num. doc'])['Fecha pedido'].min().to_frame('primer_pedido').reset_index()
ultimo_pedido = reporte3.groupby(['Fecha','Hora','Mesa','Cliente','Num. doc'])['Fecha_Hora'].max().to_frame('ultimo_pedido').reset_index()
time_pedido = pd.merge( primer_pedido,ultimo_pedido)
time_pedido['Time Estadia'] = time_pedido['ultimo_pedido']-time_pedido['primer_pedido']
time_pedido['Time_Estadia_min'] = round(time_pedido['Time Estadia'].dt.total_seconds() / 60,2)
time_pedido
```

```python
time_pedido1 = time_pedido[time_pedido['Time_Estadia_min'].isnull()]
time_pedido2 = time_pedido[~(time_pedido['Time_Estadia_min'].isnull())&(time_pedido['Time_Estadia_min']<=450)&
                           (time_pedido['Time_Estadia_min']>=0)]
time_pedido3 = time_pedido[~(time_pedido['Time_Estadia_min'].isnull())&(time_pedido['Time_Estadia_min']>450)]
time_pedido4 = time_pedido[~(time_pedido['Time_Estadia_min'].isnull())&(time_pedido['Time_Estadia_min']<0)]
```

```python
time_pedido3.drop(columns=['Time_Estadia_min'], inplace=True)
```

```python
time_pedido4.drop(columns=['Time_Estadia_min'], inplace=True)
```

```python
time4 = pd.concat([time_pedido1,time_pedido3,time_pedido2,time_pedido4])
```

```python
time4.shape
```

```python
time4["Time_Estadia_min"] = time4["Time_Estadia_min"].fillna(round(time4["Time_Estadia_min"].mean(),2))
```

```python
timefinal = time4.sort_values(['Fecha','Hora'])
```
