```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import seaborn as sns
import glob
%matplotlib inline
```

```python
proyeccion = pd.read_excel('Proyección Ventas 2023 Qarbón Trimestre(1).xlsx', sheet_name=None, header=2 ,parse_dates=['Fecha'],
                          dtype={'%': 'float'})
```

```python
dfs = []
for sheet_name, df in proyeccion.items():
    df.drop(df.index[-2:], axis=0, inplace=True)
    dfs.append(df)
```

```python
proyeccion2 = pd.concat(dfs, ignore_index=True)
proyeccion2
```

```python
proyeccion2.drop(columns=['Unnamed: 0'], inplace= True)
```

```python
proyeccion2.fillna(0, inplace=True)
```

```python
proyeccion2['Desviación % Acumulada'] = proyeccion2['Desviación % Acumulada'].map('{:.2%}'.format)
proyeccion2['%'] = proyeccion2['%'].map('{:.2%}'.format)
proyeccion2['%.1'] = proyeccion2['%.1'].map('{:.2%}'.format)
```

```python
def meses(nombre):
    df =  pd.read_excel('Proyección Ventas 2023 Qarbón Trimestre 2.xlsx', sheet_name=nombre, header=1 ,parse_dates=['Fecha'],
                          dtype={'%': 'float'})
    df.dropna(subset='Fecha',inplace = True)
    df.dropna(subset='Día',inplace = True)
    df['Var % Cocina'] = df['Var % Cocina'].map('{:.2%}'.format)
    df['Var % Bar'] = df['Var % Bar'].map('{:.2%}'.format)
    return df
```

```python
abril = meses('Abril')
mayo = meses('Mayo')
junio = meses('Junio')
```

```python
seg_trime = pd.concat([abril,mayo, junio])
seg_trime.shape
```

```python
seg_trime.rename(columns={'Venta Total Proyectada':'Venta Proyectada 1er y 2do Piso S/.',
                         'Venta Cocina Proyectada':'COCINA','Venta Bar Proyectada':'BAR',
                         'Venta Total Real':'Venta Proyectada Acumulada S/.','Venta Cocina Real':'Venta Real Acum',
                         'Var % Cocina':'%','Var % Bar':'%.1','Venta Bar Real':'Venta Real 1er y 2do Piso S/.'}, inplace = True)
```

```python
seg_trime['Desviación % Acumulada'] = 0
```

```python
seg_trime['Desviación % Acumulada'] = seg_trime['Desviación % Acumulada'].map('{:.2%}'.format)
```
