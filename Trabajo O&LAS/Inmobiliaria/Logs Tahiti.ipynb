```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.options.display.max_colwidth = 200
```

```python
cmr=  pd.read_excel('Abril 02_2023/Reporte_Cmr.xlsx', parse_dates= ['FECHA DE REGISTRO','HORA DE REGISTRO','FECHA DE ULTIMA INTERACION'])
```

```python
#cmr[cmr['CLIENTE']=='EMMA ROSA DE LOS MILAGROS PIEDRA MAZA']
cmr['CLIENTE'] = cmr['CLIENTE'].str.replace('EMMA ROSA DE LOS MILAGROS PIEDRA MAZA','ENMA ROSA DE LOS MILAGROS PIEDRA MAZA')
```

```python
cmr['CLIENTE'] = cmr['CLIENTE'].str.strip().str.upper()
cmr['CLIENTE'] = cmr['CLIENTE'].str.replace('    ',' ')
cmr['CLIENTE'] = cmr['CLIENTE'].str.replace('   ',' ')
cmr['CLIENTE'] = cmr['CLIENTE'].str.replace('  ',' ')
#cmr[cmr['CLIENTE'].str.contains('        ')]
cmr.shape[0]
```

```python
df_cmr = cmr[~(cmr['CLIENTE'].duplicated(keep='first' ))]
```

```python
df_cmr.shape
```

```python
campana = pd.read_excel('Abril 02_2023/logs/Leads x Campaña.xlsx', parse_dates=['FECHA DE REGISTRO','HORA DE REGISTRO'])
```

```python
#campana[~(campana['LOTE'].isnull()) & (campana['LOTE'].duplicated(keep=False)) ].sort_values(by='LOTE')
```
